```python
# importation des  bibliothèques pandas et numpy.
import pandas as pd
import numpy as np

# lecture d'un fichier CSV contenant des données sur les pages vues d’un site de presse professionnelle 
#en utilisant la méthode read_csv de pandas et vous stockez les données dans un DataFrame appelé df.
df = pd.read_csv(r"C:\Users\fabod\OneDrive\Bureau\SQL TEMA\2006pagesvues_complete.csv")
df
```

```python
#nouvelle colonne rubrique dans le DataFrame en extrayant la première partie de l’URL de la page (après le premier /) 
#en utilisant la méthode split pour séparer les parties de l’URL par des barres obliques (/) et en sélectionnant la deuxième partie avec [1].

df['rubrique'] = df['page_url'].str.split('/', expand=True)[1]
```

```python
df['rubrique']
```

```python
df
```

```python
#nouvelle colonne rubrique1 dans le DataFrame 
#en extrayant la deuxième partie de l’URL de la page (après le deuxième /) en utilisant la même méthode que pour la colonne rubrique.

df['rubrique1'] = df['page_url'].str.split('/', expand=True)[2]
df['rubrique1']
```

```python
df
```

```python
#enregistrement duDataFrame mis à jour dans un nouveau fichier CSV en utilisant la méthode to_csv
df.to_csv(r'C:\Users\fabod\OneDrive\Bureau\SQL TEMA\tema_views_ajout_rubriques.csv')
```

```python
# Compter les occurrences de chaque valeur dans la colonne 'rubrique1'
counts = df['rubrique1'].value_counts()

# Filtrer les valeurs avec plus de 400 occurrences
xo = counts[counts > 400].index

# Mettre à jour la colonne 'rubrique1' en marquant les valeurs non retenues
df['rubrique1'] = df['rubrique1'].apply(lambda x: x if x in xo else 'xo')
```

```python
#nouvelle colonne occurrences_rubrique1 dans le DataFrame en utilisant la méthode groupby 
#pour regrouper les données par rubrique1 
# en appliquant la fonction d’agrégation count pour compter le nombre d’occurrences de chaque rubrique.
df['occurrences_rubrique1'] = df.groupby('rubrique1')['rubrique1'].transform('count')
```

```python
# la colonne rubrique renommée en typologie en utilisant la méthode rename du DataFrame.
df.rename(columns={'rubrique': 'typologie'}, inplace=True)

print(df)
```

```python
df
```

```python
#nouvelle colonne occurrences_typologie dans le DataFrame en utilisant la même méthode que pour la colonne occurrences_rubrique1 
#mais en regroupant les données par typologie.
df['occurrences_typologie'] = df.groupby('typologie')['typologie'].transform('count')
```

```python
df
```

```python
#occurrences_combinaison dans le DataFrame en utilisant la même méthode que pour les colonnes précédentes, mais en regroupant les données par combinaison de typologie et de rubrique1
df['occurrences_combinaison'] = df.groupby(['typologie', 'rubrique1'])['typologie'].transform('count')

print(df)
```
